 Gibbs Phenomena

We analyse Giibs Phenomena while approximating discontinous function.
We try to solve problem which we face while approximating a discontinous function

Where the function is discontinous it can be seen a raining(zig zag line) behavious. This behaviour of functions is known as Gibbs Phenomena.

To get the better approximcation, we increase the number of points we are taking. Therefore as we increase the terms of sines and cosines adding up the better line we get and the raining phenomena will start vanishing. This is due to the increase in the vectors we are adding so all of them will add up and sample out the perfect approximation around the discontinous points.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size':18})

dx = 0.01
L = 2*np.pi
x = np.arange(0, L+dx, dx)
n = len(x)
nquart = int(np.floor(n/4))


```

```python
f = np.zeros_like(x)
f[nquart:3*nquart] = 1
fig, ax = plt.subplots()
ax.plot(x, f, '-', color='k', linewidth=2)

```

```python
A0 = np.sum(np.ones_like(x)) * dx 
FSf = A0/2 * np.ones_like(f)

A = np.zeros(100)
B = np.zeros(100)

for k in range(0,100):
    A[k] = np.sum(f * np.cos(2*np.pi*(k)*x/L)) * dx * 2/L
    B[k] = np.sum(f * np.sin(2*np.pi*(k)*x/L)) * dx * 2/L
    FSf = FSf + A[k]*(np.cos(2*np.pi*(k)*x/L)) + B[k]*(np.sin(2*np.pi*(k)*x/L))

plt.plot(x,FSf, color='r', linewidth=2)
# plt.plot(x, f, '-', color='k', linewidth=2)
plt.show
```
